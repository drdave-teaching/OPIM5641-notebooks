Consider the following four LP formulations. Using a graphical approach in each case, determine (a) which formulation has more than one optimal solution; (b) which formulation has an unbounded solution; (c) which formulation is infeasible; and (d) which formulation has a unique optimal solution.

# Formulation 1
$Max(Z) = 3X + 7Y$

subject to:
* $2X + Y <= 6$
* $4X + 5Y <= 20$
* $2Y <= 7$
* $2X >= 7$
* $X,Y >= 0$
