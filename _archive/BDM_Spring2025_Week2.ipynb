# For loops

```python
import numpy as np
```

```python
myData = np.arange(start=0,stop=9,step=1)
```

```python
# this prints every element in np.arange(0,9,1)
for a in np.arange(start=0,stop=9,step=1):
  print(a)
```

```python
# this prints every element in np.arange(0,9,1)
for a in myData:
  print(a)
```

```python
# prompt: string split based on space and return as list

def string_split(input_string):
    """Splits a string based on spaces and returns a list of words.
    """
    return input_string.split()

# Example usage (using the provided string):
input_string = 'My professor loves guitars'
result_list = string_split(input_string)
result_list
```

```python
for a in result_list:
  print(a)
```

```python
i = 0
for a in result_list:
  print(a)
  i = i + 2
  print(i)
```

```python
for a in np.arange(start=0,stop=5,step=2):
    print('This is a:', a)
    print('----------')
    for b in np.arange(start=0,stop=5,step=1):
      print('This is b:', b)
    print('==========')
```
