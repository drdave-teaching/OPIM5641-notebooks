# Parametric MC simulation (one run)

```python


# a place to store results
results = []

# initialize the start points (initital investment, contribution)
initialAmount = 100000
contribution = 10000 

# for loop (30 years)
for a in np.arange(0,30,1):
  print(a)
  # define a rate (a random number from a normal distribution)
  rate = 1+ np.random.normal(loc=7.8, scale=19, size=1)/100
  print(rate)

  # apply the rate to savings 
  savings = (initialAmount+contribution)*(rate)
  print(savings)

  # add the value to the list
  results.append(savings)
  # update the initial amount (information handshake)
  initialAmount = savings 

results = pd.DataFrame(results)
results.plot()
plt.show()
```
