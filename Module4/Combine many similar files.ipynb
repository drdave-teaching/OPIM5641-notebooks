You will need to mount your Drive for this one! So long as files have the same extension and format (same columns, regardless of number of rows), you can merge them into one final dataframe. If they are HUGE files, there are some low-memory options that you can utilize.

* https://stackoverflow.com/questions/20906474/import-multiple-csv-files-into-pandas-and-concatenate-into-one-dataframe

```python
import pandas as pd
import glob

path = '/content/drive/MyDrive/Online Materials/Prep/Module4_MC_Nonlinear/Week1_Introduction/Webscraping and Geocoding Addresses/master_All_CVS_locations' # use your path on your Drive
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
all_files
```

Then after everything is merged, you can clean up in one fell swoop and you are ready to rock!

```python
frame
```

```python
df = frame
```

## Drop Missing Values

```python
print('Before:', df.shape)
df.dropna(inplace=True)
print('After:', df.shape)
```

## Delete a column


```python
print('Before:', df.shape)
del df['Unnamed: 0']
print('After:', df.shape)
```

## Drop Duplicates

```python
print('Before:', df.shape)
df.drop_duplicates(inplace=True)
print('After:', df.shape)
```

## Make a map

```python
import folium as folium
map1 = folium.Map(
    location=[30, -86], # lat lon of alabama!
    tiles='cartodbpositron',
    zoom_start=1,
)
df.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map1), axis=1)
map1
```
