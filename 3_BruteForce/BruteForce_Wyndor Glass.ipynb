# Problem Description
The WYNDOR GLASS CO. produces high-quality glass products, including windows and glass doors. It has three plants. Aluminum frames and hardware are made in Plant 1, wood frames are made in Plant 2, and Plant 3 produces the glass and assembles to products.

Due to declining earnings, top management has decided to revamp the company's product line. Unprofitable products are being discontinued - releasing production capacity to launch two new products having large sales potential.

* Product 1: An 8-foot glass door with aluminum framing
* Product 2: A 4x6 foot double-hung wood frame window

## Some constraints...

Product 1 requires some of the production capacity in Plants 1 and 3, but none in Plant 2. Product 2 needs only Plants 2 and 3. The marketing division has concluded that the company could sell as much product as could be produced by these plants.  However, because both products would be competing for the same production capacity in Plant 3, it is not clear which *mix* of the two products would be the *most profitable*.

Therefore, an Operations Research (OR) team has been formed to study this question.

## A meeting with management...
After meeting with management, the problem has more concretely been defined as:

* Determine what the production rates should be for the two products in order to maximize their total profit, subject to the restrictions imposed by the limited production capacities avialble in the three plants. Each product will be produced in batches of 20, so the production rate is defined as the number of batches produced per week.

* Because the work on a current batch of a particular product is commonly partially completed by the end of the week,  the production rate can be an integer (1,2,3 batches...) or non-integer (0.5 batches).

* Any combination of production rates that satisfies the restrictions imposed by the limited production capacities is permitted, including none of one product or as much as possible of the other.

## Formulation as a linear program
The definition of the problem given above indicates that the decisions to be made are the number of batches of the respective products to be produced per week so as to maximize their total profit.

You can start to craft this as a linear program even though you only have the brute force method to help you!

### Decision Variables and Objective Function
* $x_1$ = number of batches of product 1 produced per week
* $x_2$ = number of batches of product 2 produced per week
* $Z$ = total profit per week in thousands of dollars from producing these two products

Thus, $x_1$ and $x_2$ are the decision variables for the model. Hence, we obtain the objective function.

$Z = 3000x_1 + 5000x_2$

The objective is to select the values of $x_1$ and $x_2$ to maximize $Z$ subject to restrictions imposed on their values by the limited production capacities available in the three plants.

### Constraints
The time constraints can be represented as inequalities. The table above indicates that each batch of Product 1 uses one hour of production time in Plant 1 and that only 4 hours per week are available. This can be represented as $x_1 \leq 4 $. Similarly, Plant 2 imposes $2x_2 \leq 12$ and Plant 3 imposes $3x_1 + 2x_2 \leq 18$.

### Summarize as an LP!
The problem is to select values of $x_1$ and $x_2$ so as to:

Maximize $Z = 3000x_1 + 5000x_2$

subject to the restrictions (constraints):
* $x_1 \leq 4 $
* $2x_2 \leq 12$
* $3x_1 + 2x_2 \leq 18$

Notice how all of the data has been translated from a word problem, to a table and now to this mathematical representation!

### A resource-allocation problem
This problem is a classic example of a **resource-allocation problem**, the most common type of linear programming problem. The key characteristic of resource-allocation problems is that most or all of their functional constraints are *resource constraints*.

The right-hand side (RHS) of a resource constraint represents the amount available of some resource, and the left-hand side (LHS) represents the amount that is used of that resource, so the LHS must $\leq$ the RHS.

Product mix probables are one type of resource-allocation problems.

# Brute Force
For this initial solving of the problem, we will simply enumerate EVERY SINGLE POSSIBLE FEASIBLE COMBINATION OF VALUES! We will select the values of $x_1$ and $x_2$ that maximizes the objective function.

* A **feasible solution** is a solution for which all the constraints are satisfied
* An **infeasible solution** is a solution for which at least one constraint is violated
* An **optimal solution** is a feasible solution that has the *most favorable value* of the objective function

If it's a maximization problem, the most favorable solution is the largest optimal solution (think big profits!) If it's a minimization problem, the most favorable solution is the smallest number (think low cost!) Most problems will have just one optimal solution  (but sometimes there can be more than one... more on that later.)

# Something new... no `continue` statements
Use the `and` statement to get it to work in a more logical way!

`continue` is great, but it can look a little counterintuitive.

```python
import numpy as np

# Variables containing current best value and current best solution
best_profit = 0
best_prod1 = 0
best_prod2 = 0

for prod1 in np.arange(1, 41, 1):
    for prod2 in np.arange(1, 41, 1):
        plant1_hours = plant1["prod1"] * prod1 + plant1["prod2"] * prod2
        plant2_hours = plant2["prod1"] * prod1 + plant2["prod2"] * prod2
        plant3_hours = plant3["prod1"] * prod1 + plant3["prod2"] * prod2
        # one if statement with all constraints/conditions
        if plant1_hours <= total_time["plant1"] and
          plant2_hours <= total_time["plant2"] and
          plant3_hours <= total_time["plant3"]:

          # if all of that is true, print this
          print("Feasible solution:",  prod1, ',', prod2)
          # and check to see if current_profit is better than best_profit
            current_profit = profit["prod1"] * prod1 + profit["prod2"] * prod2

          # if current_profit is better... store profit and prod mix
            if current_profit > best_profit:
                best_profit = current_profit
                best_prod1 = prod1
                best_prod2 = prod2
                print('Profit:', best_profit)

print("Maximum profit:", best_profit)
print("prod1", best_prod1)
print("prod2", best_prod2)
```
