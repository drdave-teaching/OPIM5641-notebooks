## Loop control
Continue, break

### break: stop the for loop entirely

```python
# spice it up! v1
# a blank [list] to store results
resultDF = []

for a in myList:
  print("this is a:", a)

  # conditional/loop control
  # if current value is exactly equal to 3... then break the loop! stop everything!
  if a == 3:
    break

  # assign output to tmp
  tmp = a*2
  print('---this is tmp:', tmp)

  # save it to my list one at at time
  resultDF.append(tmp)
  print('this is what resultDF looks like:', resultDF, '\n')
```

### continue: continue onto next value in loop

```python
# spice it up! v1
# a blank [list] to store results
resultDF = []

for a in myList:
  print("this is a:", a)

  # conditional/loop control
  # if current value is exactly equal to 3... then break the loop! stop everything!
  if a == 3:
    print('---')
    print("CONTINUE ONTO NEXT VALUE, IT'S QUAL TO 3!!!")
    continue

  # assign output to tmp
  tmp = a*2
  print('---this is tmp:', tmp)

  # save it to my list one at at time
  resultDF.append(tmp)
  print('this is what resultDF looks like:', resultDF, '\n')
```

# Retirement

```python
import numpy as np
years_working = 35
start_salary = 100000
my_contrib = 0.05
their_contrib = 0.07
my_raise = 0.03
#stock_growth = 0.07
stock_growth = np.random.normal(loc=0.07, scale=0.2)
```

```python
print(stock_growth)
```

## Get it right once for 35 years

```python
import numpy as np
years = np.arange(0, years_working, 1) # Use years_working for the loop
resultDF = []

# initialize current money
current_money = 0 # Start with 0 current savings
for a in years:
  print('---')
  print('Year:', a)
  # Apply the raise to the salary for the year
  salary_this_year = start_salary * (1 + my_raise)**a # Calculate salary for this year
  print('My salary this year:', salary_this_year)
  contrib = salary_this_year * (my_contrib + their_contrib)
  print('I will add this much money:', contrib)
  current_money = contrib + current_money
  # stock market growth
  stock_growth = np.random.normal(loc=0.07, scale=0.2)
  print('stock growth:', stock_growth)
  current_money = current_money * (1+stock_growth)
  print('At the end of the year I will have:', current_money)
  # Append current_money to resultDF in each iteration
  resultDF.append(current_money)
```

```python
resultDF
```

```python
import matplotlib.pyplot as plt
import pandas as pd
resultDF = pd.DataFrame(resultDF)
resultDF.head()
```

```python
resultDF = pd.DataFrame(resultDF)
resultDF.plot()
plt.ylim([0, resultDF.max().max() * 1.1]) # Adjust y-limit dynamically
plt.show()
```

```python
import numpy as np
years = np.arange(0, 35, 1) # Use years_working for the loop
resultDF = []

# initialize current money

for b in np.arange(0,10000,1):
  current_money = 0 # Start with 0 current savings
  for a in years:
    print('---')
    print('Year:', a)
    # Apply the raise to the salary for the year
    salary_this_year = start_salary * (1 + my_raise)**a # Calculate salary for this year
    print('My salary this year:', salary_this_year)
    contrib = salary_this_year * (my_contrib + their_contrib)
    print('I will add this much money:', contrib)
    current_money = contrib + current_money
    # stock market growth
    stock_growth = np.random.normal(loc=0.07, scale=0.2)
    print('stock growth:', stock_growth)
    current_money = current_money * (1+stock_growth)
    print('At the end of the year I will have:', current_money)
    # Append current_money to resultDF in each iteration
    if a ==34:
      resultDF.append(current_money)
print('BURRRRP... ALL DONE!')
print(np.shape(resultDF))
```

```python
resultDF = pd.DataFrame(resultDF)
resultDF.hist(bins=100)
plt.title('Distribution of Expected Retirement Savings')
plt.ylabel("Frequency")
plt.xlabel('Savings $')
plt.axvline(np.mean(resultDF), color='r', linestyle='dashed', linewidth=3, label=f'Mean: {np.mean(resultDF):,.0f}')
plt.axvline(np.quantile(resultDF, 0.25), color='black', linestyle='dashed', linewidth=1, label=f'25th perc: {np.quantile(resultDF, 0.25):,.0f}')
plt.axvline(np.quantile(resultDF, 0.75), color='black', linestyle='dashed', linewidth=1, label=f'75th perc: {np.quantile(resultDF, 0.75):,.0f}')
plt.legend()
plt.show()
```

# Spaghetti

```python
import numpy as np
years = np.arange(0, 35, 1) # Use years_working for the loop
finalDF = pd.DataFrame()
# initialize current money

for b in np.arange(0,10000,1):
  resultDF = [] # Reset resultDF for each simulation
  current_money = 0 # Start with 0 current savings
  for a in years:
    # print('---')
    # print('Year:', a)
    # Apply the raise to the salary for the year
    salary_this_year = start_salary * (1 + my_raise)**a # Calculate salary for this year
    # print('My salary this year:', salary_this_year)
    contrib = salary_this_year * (my_contrib + their_contrib)
    # print('I will add this much money:', contrib)
    current_money = contrib + current_money
    # stock market growth
    stock_growth = np.random.normal(loc=0.07, scale=0.2)
    # print('stock growth:', stock_growth)
    current_money = current_money * (1+stock_growth)
    # print('At the end of the year I will have:', current_money)
    # Append current_money to resultDF in each iteration
    resultDF.append(current_money)
  resultDF = pd.DataFrame(resultDF, columns=[f'Simulation_{b}']) # Create DataFrame with unique column name
  finalDF = pd.concat([finalDF, resultDF], axis=1) # Use pd.concat()
print('BURRRRP... ALL DONE!')
print(np.shape(finalDF))
```
