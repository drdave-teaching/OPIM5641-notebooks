# Solve this example on MC for Cell Phones

**[Profit Analysis 14.1 - Powell]** A consumer electronics firm produces a line
of battery rechargers for cell phones. The following distributions apply:

* Unit price triangular with a minimum of 18.95 USD, most
likely value of 24.95 USD, and maximum of 26.95 USD
* Unit cost uniform with a minimum of 12.00 USD and a
maximum of 15.00
* Quantity sold = $10,000 – 250*UnitPrice$, plus a random term
given by a normal distribution with a mean of
0 and a standard deviation of 10
* Fixed costs normal with a mean of 30,000 USD and a standard deviation of 5,000 USD

1. What is the expected profit?
2. What is the probability of a loss?
3. What is the maximum loss?

```python
import numpy as np
unitPrice = np.random.triangular(left=18.95, mode= 24.95, right=26.95)
unitCost = np.random.uniform(low=12, high=15)
qtySold = 10000 - 250*unitPrice + np.random.normal(loc=0, scale=10)
fixedCosts = np.random.normal(loc=30000, scale= 5000)
revenue = qtySold*unitPrice
variableCosts = qtySold*unitCost
profit = revenue - fixedCosts - variableCosts
print(profit)
```

```python
results = []
for a in np.arange(0,10000,1):
  import numpy as np
  unitPrice = np.random.triangular(left=18.95, mode= 24.95, right=26.95)
  unitCost = np.random.uniform(low=12, high=15)
  qtySold = 10000 - 250*unitPrice + np.random.normal(loc=0, scale=10)
  fixedCosts = np.random.normal(loc=30000, scale= 5000)
  revenue = qtySold*unitPrice
  variableCosts = qtySold*unitCost
  profit = revenue - fixedCosts - variableCosts
  #print(profit)
  results.append(profit)
print(np.shape(results))
```

## Example - Furniture (extracted from the textbook)

**Problem Description:**
Veerman Furniture Company makes three kinds of office furniture: chairs, desks, and tables. Each product requires some labor in the parts fabrication department, the assembly department, and the shipping department. The furniture is sold through a regional distributor, who has estimated the maximum potential sales for each product in the coming quarter. Finally,the accounting department has provided some data showing the profit contributions on each product. The decision problem is to determine the product mix - that is, to **maximize Veerman’s profit for the quarter by choosing production quantities for chairs, desks, and tables**.

The following data summarizes the parameters of the problem:

Department | Chairs | Desks | Tables | Hours Available
--- | --- | --- | --- | ---
Fabrication | 4 | 6 | 2 | 1,850
Assembly | 3 | 5 | 7 | 2,400
Shipping | 3 | 2 | 4 | 1,500
Demand Potential | 360 | 300 | 100 |
Profit (USD) | 15 | 24 | 18 |

```python
## BRUTE FORCE ALGORITHM: iterate over EVERY combination
# 360*300*100 combinations...
# Benefit: this model will be 100% interpretable
# Con: it will take a long time... with more variables and constraints... even longer

# decision variables: C, D, T
# one nested for loop for each decision

# initialize placeholders
bestProfit = 0
bestC = 0
bestD = 0
bestT = 0

# brute force algo
for c in np.arange(0,360+1,1):
  for d in np.arange(0,300+1,1):
    for t in np.arange(0,100+1,1):
      #print(c,d,t)
      # make sure all cosntraints are valid, and if they are...
      if(
          (4*c + 6*d + 2*t <= 1850) & # fabrication
          (3*c + 5*d + 7*t <= 2400) & # assembly
          (3*c + 2*d + 4*t <= 1500) & # shipping
          (c <= 360) &
          (d <= 300) &
          (t <= 100) &
          (c >= d+t) # have at least one chair for each desk and table
      ):
        #print("Valid!")
        # calculate the currentProfit
        currentProfit = 15*c + 24*d + 18*t
        if currentProfit > bestProfit:
          bestProfit = currentProfit
          bestC = c
          bestD = d
          bestT = t
      else:
        #print('Invalid!')
        continue # go to the next spin of the loop
print('ALL DONE!')
print('bestProfit:', bestProfit)
print('bestC:', bestC)
print('bestD:', bestD)
print('bestT:', bestT)
```
